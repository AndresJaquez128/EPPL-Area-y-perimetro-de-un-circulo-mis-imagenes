## formula para calcular area
$A=\pi*r^2$

---
## Formula para calcular perimetro
$P=2*\pi*r$

---
## Datos de entrada
$r$  ?

```python
# Calcular area y perimetro
# Variables A, P, r
# Formulas A = pi*r^2 y P = 2 * pi * r
# Calcular el valor de r
import math
print("Calcular el area y perimetro de un circulo")
r = float(input("Ingrese el radio del circulo: "))
A = math.pi * r**2
P = 2 * math.pi * r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Jaquez Camacho Joel andres NL 27")
```
